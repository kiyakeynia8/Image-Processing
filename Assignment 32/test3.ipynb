```python
import cv2
import matplotlib.pyplot as plt
```

```python
image_low_contrast = cv2.imread("input/4.png",0)
```

```python
hist = cv2.calcHist([image_low_contrast], [0], None, [256], [0, 256])
```

```python
plt.plot(hist)
plt.show()
```

```python
image_high_contrast = cv2.equalizeHist(image_low_contrast)

cv2.imwrite("output assignment_32/test3.png", image_high_contrast)
```

```python
hist = cv2.calcHist([image_high_contrast], [0], None, [256], [0, 256])
plt.plot(hist)
plt.show()
```
